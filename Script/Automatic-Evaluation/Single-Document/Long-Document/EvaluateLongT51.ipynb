CHRF

```python
!pip install sacrebleu sacrebleu
```

```python
results=[]
```

```python
chrf = evaluate.load("chrf")
resultsc1 = chrf.compute(predictions=[M6], references=[reference],beta=1,word_order=0)
print(resultsc1)
results.append(resultsc1)
```

```python
chrf = evaluate.load("chrf")
resultsc2 = chrf.compute(predictions=[M7], references=[reference],beta=1,word_order=0)
print(resultsc2)
results.append(resultsc2)
```

```python
final_results=[]
for i in range(2):
  tmp=results[i]
 
  val=tmp["score"]/100
  final_results.append(round(val,4))



final_results
```

```python
with open("/content/drive/MyDrive/Colab_Notebooks/CHFR/resultsLongT5_1.txt", 'w') as f:
  f.write(f'{final_results}')
```
